```python
df['words'] = df['response_text'].apply(lambda x: x.split())
df.head()
```

```python
df = df.explode('words')
df['words'] = df['words'].str.strip()
df.head()
```

```python
import string
df['words'] = df['words'].str.strip(string.punctuation)
df['words'] = df['words'].str.lower()
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction import text

nltk.download('stopwords')
nltk_stopwords = stopwords.words('english')
sklearn_stopwords = list(text.ENGLISH_STOP_WORDS)
all_stopwords = list(set(nltk_stopwords + sklearn_stopwords))
```

```python
print(len(nltk_stopwords))
print(len(sklearn_stopwords))
print(len(all_stopwords))
```

```python
df.head()
```

```python
df['is_stopword'] = df['words'].apply(lambda x: str(x)  in all_stopwords)
df.head()
```

```python
df = df[df['is_stopword'] == False ] 
df
```

```python
df = df.drop_duplicates(['response_id', 'words'])
len(df)
```

```python
df_flagged = df[df['class'] == 'flagged']
flagged_word_counts = df_flagged.groupby('words').count().iloc[:, 0:1]
flagged_word_counts = flagged_word_counts.sort_values('response_id', ascending=False)
flagged_word_counts.head(25)
```

```python
df_not_flagged = df[df['class'] == 'not_flagged']
not_flagged_word_counts = df_not_flagged.groupby('words').count().iloc[:, 0:1]
not_flagged_word_counts = not_flagged_word_counts.sort_values('response_id', ascending=False)
not_flagged_word_counts.head(25)
```

```python
flagged_top_words = flagged_word_counts.index[0:30]
not_flagged_top_words = not_flagged_word_counts.index[0:30]
flagged_top_words
```

```python
common_words = set(flagged_top_words).intersection(not_flagged_top_words)
common_words
```

```python
flagged_top_words_filtered = [ word for word in flagged_top_words if word not in common_words ]
```
