```python
from datetime import date
def calculate_age(built_year):
    today = date.today().year
    age = today - built_year
    return age
df['age'] = df['YearBuilt'].apply(calculate_age)
df.head()
```

```python
list(set(df['Neighborhood'].values))
from sklearn.preprocessing import OneHotEncoder
#fit and transform--> here it works as if to represent functions in the mathematics
```

```python
enc = OneHotEncoder(handle_unknown = 'ignore')
enc.fit(df[['Neighborhood']])
```

```python
ohe_array = enc.transform(df[['Neighborhood']]).toarray()
one_hot_df = pd.DataFrame(ohe_array, columns = enc.get_feature_names_out())
one_hot_df
```

```python
df = df.drop(['Neighborhood'], axis = 1)
df
```

```python
print(len(df))
print(len(one_hot_df))
```

```python
new_df = pd.concat([df, one_hot_df], axis = 1, ignore_index=False)
new_df
```

```python
from sklearn.linear_model import LinearRegression
X = new_df.drop(columns = ['index', 'Price'])
y = new_df['Price']
```

```python
## Let us standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
# X = scaler.transform(X)
X = pd.DataFrame(scaler.transform(X),columns = X.columns)
```

```python
print(X.mean(axis=0))
print("-"*50)
print(X.std(axis=0))
```

```python
reg = LinearRegression().fit(X, y)
pred = reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, pred)
#mean absolue error, mean square error, root mean square error, mean absolute percentage error
```

```python
reg.score(X, y)
```

```python
reg.coef_
```

```python
import matplotlib.pyplot as plt
plt.bar(reg.feature_names_in_, reg.coef_)
plt.xlabel('Features')
plt.ylabel('Feature importance')
plt.xticks(reg.feature_names_in_, rotation=45)
```

```python
from sklearn.preprocessing import OneHotEncoder

def encode_cat_feture(curr_feature):
    enc = OneHotEncoder(handle_unknown='ignore')
    enc.fit(df[[curr_feature]])
    ohe_array = enc.transform(df[[curr_feature]]).toarray()
    one_hot_df = pd.DataFrame(ohe_array, columns=enc.get_feature_names_out())
    return one_hot_df
categorical_features = ["person_gender", "person_education", "person_home_ownership", "loan_intent", "previous_loan_defaults_on_file"]
```

```python
ohe_dfs = []
for feature in categorical_features:
    ohe_dfs.append(encode_cat_feture(feature))
```

```python
ohe_dfs[4]
```

```python
for feature in categorical_features:
    df = df.drop([feature], axis=1)
df.head()
```

```python
df
```

```python
ohe_dfs_concat = pd.concat(ohe_dfs, axis=1, ignore_index=False)
new_df = pd.concat([df, ohe_dfs_concat], axis=1, ignore_index=False)
new_df = new_df.dropna()
corr = df.corr()
corr
```

```python
import numpy as np
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# # Generate a custom diverging colormap
# cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap='Blues', center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
